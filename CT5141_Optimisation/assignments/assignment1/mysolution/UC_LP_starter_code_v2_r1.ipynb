## 1. Introduction

There are 12 electricity generators of four types. Each generator has a cost for producing (per MW) and produces a certain amount of CO2 (per MV). 

We want to solve two problems: 
* We want to minimise total cost.
* We want to minimise total emissions.

## 2. Decision Variables

Let $X_{ij}$ be the amount of electrivity generated by generator $i$ at time $j$, where $i=\{ 0, 2, 3, ..., 11 \}$, $j=\{0, 2, 3, ..., 23 \}$.

## 3. Formalising the objective

### 3.1 Minimise total cost

$$
\sum_{j=0}^{23}(\sum_{i=0}^{3}1.4\times X_{ij}+\sum_{i=4}^{6}4.4\times X_{ij}+9.1\times X_{7_j}+9.5\times X_{8_j}+\sum_{i=9}^{11}5.4\times X_{ij})
$$

### 3.2 Minimise total CO2 emissions

$$
\sum_{j=0}^{23}(\sum_{i=0}^{3}0.024\times X_{ij}+\sum_{i=4}^{6}0.82\times X_{ij}+0.49\times X_{7_j}+0.51\times X_{8_j}+\sum_{i=9}^{11}0.0\times X_{ij})
$$

## 5. Linear programming

In this case, we use linear programming to solve this pronlem. Because: (1) the search space is consist of real-value decision variables, (2) the objective can be formalised as linear equation in decision variables and (3) the constrains can be expressed as linear equations and inequalities in decision variables. 

## 6. Solve problrm by coding

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from ortools.linear_solver import pywraplp


generator_info = pd.read_csv("generator_info.csv", delimiter=",")
demand = np.loadtxt("demand.csv")
solar_curve = np.loadtxt("solar_curve.csv")
ngenerators = len(generator_info["name"])
nhours = 24 # hard-coded as unlikely to change
    
CO2 = generator_info["CO2 (tons/MW)"].values
cost = generator_info["cost (EUR/MW)"].values
```

```python
# Lower and upper bounds of electricity production on each generator
LB = [10.0, 10.0, 10.0, 1.0, 100.0, 100.0, 10.0, 100.0, 50.0, 0.0, 0.0, 0.0] 
UB = [100.0, 80.0, 60.0, 10.0, 900.0, 600.0, 100.0, 400.0, 300.0, 20.0, 60.0, 30.0]

s = [0.0, 0.5, 1.0, 0.5, 0.0] # solar generator supply ratio
```

### 6.1 Minimise total cost.

```python
# Instantiate an LP solver using the GLOP solver
solver_mincost = pywraplp.Solver.CreateSolver('GLOP')
```

```python
# We store these DVs in an 2D-list, which i represents index of generator 
# j represents index of time.
X = [[0 for j in range(nhours)] for i in range(ngenerators)]
```

```python
# Create decision variables with their lower and upper bound.

# We assign the DVs which are not solar generators to the 2D-list
for j in range(nhours):
    for i in range(9):
        X[i][j] = solver_mincost.NumVar(LB[i], UB[i], f"X{i}_{j}") 
```

```python
# We assign the DVs which are solar generators to the 2D-list
# Sorlar generator varies by hour.
for i in range(9, ngenerators):
    for j in range(0, 6):
        X[i][j] = solver_mincost.NumVar(LB[i], UB[i]*s[0], f"X{i}_{j}") 
    for j in range(6, 11):
        X[i][j] = solver_mincost.NumVar(LB[i], UB[i]*s[1], f"X{i}_{j}") 
    for j in range(11, 16):
        X[i][j] = solver_mincost.NumVar(LB[i], UB[i]*s[2], f"X{i}_{j}") 
    for j in range(16, 19):
        X[i][j] = solver_mincost.NumVar(LB[i], UB[i]*s[3], f"X{i}_{j}") 
    for j in range(19, nhours):
        X[i][j] = solver_mincost.NumVar(LB[i], UB[i]*s[4], f"X{i}_{j}") 
```

```python
# Add further constraints
# The demand from electricity customers varied by hour of the day
solver_mincost.Add(sum(X[i][0] for i in range(ngenerators)) == 1461)
solver_mincost.Add(sum(X[i][1] for i in range(ngenerators)) == 1446)
solver_mincost.Add(sum(X[i][2] for i in range(ngenerators)) == 1446)
solver_mincost.Add(sum(X[i][3] for i in range(ngenerators)) == 1438)
solver_mincost.Add(sum(X[i][4] for i in range(ngenerators)) == 1425)
solver_mincost.Add(sum(X[i][5] for i in range(ngenerators)) == 1420)
solver_mincost.Add(sum(X[i][6] for i in range(ngenerators)) == 1467)
solver_mincost.Add(sum(X[i][7] for i in range(ngenerators)) == 1503)
solver_mincost.Add(sum(X[i][8] for i in range(ngenerators)) == 1507)
solver_mincost.Add(sum(X[i][9] for i in range(ngenerators)) == 1504)
solver_mincost.Add(sum(X[i][10] for i in range(ngenerators)) == 1504)
solver_mincost.Add(sum(X[i][11] for i in range(ngenerators)) == 1502)
solver_mincost.Add(sum(X[i][12] for i in range(ngenerators)) == 1498)
solver_mincost.Add(sum(X[i][13] for i in range(ngenerators)) == 1496)
solver_mincost.Add(sum(X[i][14] for i in range(ngenerators)) == 1492)
solver_mincost.Add(sum(X[i][15] for i in range(ngenerators)) == 1492)
solver_mincost.Add(sum(X[i][16] for i in range(ngenerators)) == 1492)
solver_mincost.Add(sum(X[i][17] for i in range(ngenerators)) == 1509)
solver_mincost.Add(sum(X[i][18] for i in range(ngenerators)) == 1518)
solver_mincost.Add(sum(X[i][19] for i in range(ngenerators)) == 1517)
solver_mincost.Add(sum(X[i][20] for i in range(ngenerators)) == 1515)
solver_mincost.Add(sum(X[i][21] for i in range(ngenerators)) == 1490)
solver_mincost.Add(sum(X[i][22] for i in range(ngenerators)) == 1479)
solver_mincost.Add(sum(X[i][23] for i in range(ngenerators)) == 1463)
```

```python
# Create the objective function
# minimise total cost
solver_mincost.Minimize(
    sum(cost[i] * X[i][j] for j in range(nhours) for i in range(ngenerators))
) 
```

```python
# Create an array to store the solution value for each generator i and time j.
names_c = []
sol_values_c = []
costs_c = []
```

```python
# Solve the system
result = solver_mincost.Solve()

if result == solver_mincost.OPTIMAL:
    print("The solution to minimize the cost. \n")
    print("Xi_j represents we use generator i at time j \n")
    
    for v in solver_mincost.variables():
        names_c.append(v.name())
        sol_values_c.append(v.solution_value())
        costs_c.append(solver_mincost.Objective().GetCoefficient(v))
        
        #print(f"{v.name()} = {v.solution_value():.5}; cost {solver_mincost.Objective().GetCoefficient(v):.5}")
        
    print(f"Optimal objective value = {solver_mincost.Objective().Value():.5}")
else:
    print("Problem is not feasible")
```
