```python
import networkx as nx

def build_graph(dict_f):
    
    G = nx.Graph()
    
    for society in dict_f:
        for i in range(0, len(dict_f[society])-1):
            for j in range(i+1, len(dict_f[society])):
                G.add_edge(dict_f[society][i],dict_f[society][j],attr=society)
    
    print("Plot the graph:")
    nx.draw_networkx(G)
    
    return G
```

```python
G = build_graph(dict_f1)
```

```python
def find_path(G, start, end, cutoff=4):
    # find all paths which length <= 4
    paths = nx.all_simple_paths(G, source=start, target=end, cutoff=cutoff)
    
    if list(paths) == []:
        # if there's no path between the start node and end node
        print(f"No path between {start} and {end}!")
        
    else:
        paths = nx.all_simple_paths(G, source=start, target=end, cutoff=cutoff)
        print(f"""All the paths from {start} to {end} with the paths of length <= {cutoff}:
        {list(paths)}""")
       
        # find the shortest path
        shortest_path = nx.shortest_path(G, "Dracula", "Pumpkin")
        print(f"\nThe shortest path from {start} to {end} is {shortest_path}")
```

```python
find_path(G, "Dracula", "Pumpkin", cutoff=5)
```
