```python
test = {
"Karate Society": ["Dracula", "Frankenstein", "Werewolf", "Pilot", "Luke", "Leia"],
"Debating Society": ["Dracula", "Frankenstein", "Leia", "Cockroach", "Iron Man", "Zombie"],
"Musical Society": ["Bunny", "Pumpkin", "Cinderella", "Goblin Queen", "Goblin King"],
"History Society": ["Bunny", "Pilot", "Luke"]
}
```

```python
import networkx as nx

G = nx.Graph()
```

```python
for society in test:
    for i in range(0, len(test[society])-1):
        for j in range(i+1, len(test[society])):
            G.add_edge(test[society][i],test[society][j],attr=society)
```

```python
paths = nx.all_simple_paths(G, source="Dracula", target="Pumpkin", cutoff=4)
print(list(paths))
for path in paths:
    print(len(path))
```

```python
shortest_path = nx.shortest_path(G, "Dracula", "Pumpkin")
```
